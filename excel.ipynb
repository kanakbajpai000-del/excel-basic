 1.Explain the difference between Relative , Absolute and Mixed Cell Referencing.
 ->**Relative referencing**-changes when copied,by default it adjust to new cell reference.
 eg.- if formula for cell 1 is A1+B1
      then for cell 2 it becomes A2+B2

 ->**Absolute referencing**- Uses dollar signs along with both column and row eg.-$ , always same for each row and column.
      eg.- for cell 1 is A1+$B$1
      then cell 2 is A2 +$B$1

 -> ** Mixed referencing**- Uses dollar with either row or column , by fixing one of them that one is always remain same but other one is changed for different cell according to adjusment.
     eg.- for cell 1 is A1+ $B1
      for cell 2 is A1+ $B2
